```python
# -------------------------------
# 📦 INSTALL (run once if needed)
# -------------------------------
# !pip install SpeechRecognition gTTS scikit-learn nltk

import nltk
import string
import numpy as np

# Optional (voice)
try:
    import speech_recognition as sr
    VOICE_AVAILABLE = True
except:
    VOICE_AVAILABLE = False

try:
    from gtts import gTTS
    from IPython.display import Audio, display
    SPEAK_AVAILABLE = True
except:
    SPEAK_AVAILABLE = False

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('punkt')

# -------------------------------
# 📚 KNOWLEDGE BASE
# -------------------------------
# -------------------------------
# 📚 KNOWLEDGE BASE
# -------------------------------
faq_data = {}   # ✅ ADD THIS LINE
faq_data.update({

"how to apply for university admission":
"First you fill online admission form on university website.\nسب سے پہلے آپ یونیورسٹی کی ویب سائٹ پر آن لائن فارم بھرتے ہیں۔",

"admission process kya hota hai":
"Admission process includes form submission, test, and merit list.\nایڈمیشن پراسیس میں فارم، ٹیسٹ اور میرٹ لسٹ شامل ہوتے ہیں۔",

"what documents are required for admission":
"You need CNIC/B-Form, result cards, and photos.\nآپ کو شناختی کارڈ یا ب فارم، رزلٹ کارڈ اور تصاویر چاہیے ہوتی ہیں۔",

"what is admission form":
"Admission form is used to apply for university.\nایڈمیشن فارم یونیورسٹی میں داخلہ لینے کے لیے ہوتا ہے۔",

"how to fill admission form":
"Enter your personal and academic details carefully.\nاپنی ذاتی اور تعلیمی معلومات احتیاط سے درج کریں۔",

"what is entry test":
"Entry test checks your knowledge for admission.\nانٹری ٹیسٹ آپ کی قابلیت چیک کرتا ہے۔",

"entry test kya hota hai":
"Entry test admission ke liye zaroori hota hai.\nانٹری ٹیسٹ داخلے کے لیے ضروری ہوتا ہے۔",

"what is merit list":
"Merit list shows selected students based on marks.\nمیرٹ لسٹ منتخب طلباء کو ظاہر کرتی ہے۔",

"merit list kab lagti hai":
"Merit list admission deadline ke baad lagti hai.\nمیرٹ لسٹ آخری تاریخ کے بعد لگتی ہے۔",

"how to check merit list":
"You can check it on university website.\nآپ اسے یونیورسٹی کی ویب سائٹ پر دیکھ سکتے ہیں۔",

"what is eligibility criteria":
"Eligibility means required marks and qualifications.\nاہلیت کا مطلب ضروری نمبرز اور قابلیت ہے۔",

"minimum marks required for admission":
"Usually 50% to 60% marks required.\nعام طور پر 50 سے 60 فیصد نمبرز ضروری ہوتے ہیں۔",

"can i apply in multiple universities":
"Yes, you can apply in multiple universities.\nجی ہاں، آپ کئی یونیورسٹیوں میں اپلائی کر سکتے ہیں۔",

"what is admission fee":
"Admission fee is first payment after selection.\nایڈمیشن فیس سلیکشن کے بعد پہلی ادائیگی ہوتی ہے۔",

"how to pay admission fee":
"You can pay through bank or online.\nآپ بینک یا آن لائن ادائیگی کر سکتے ہیں۔",

"what happens after selection":
"You confirm seat by paying fee.\nآپ فیس دے کر اپنی سیٹ کنفرم کرتے ہیں۔",

"what if i miss deadline":
"You may lose your admission chance.\nآپ کا داخلہ ضائع ہو سکتا ہے۔",

"can i change program after admission":
"Some universities allow program change.\nکچھ یونیورسٹیاں پروگرام تبدیل کرنے دیتی ہیں۔",

"what is waiting list":
"Waiting list includes backup candidates.\nویٹنگ لسٹ میں اضافی امیدوار ہوتے ہیں۔",

"how many merit lists are announced":
"Usually 2 to 3 merit lists are announced.\nعام طور پر 2 سے 3 میرٹ لسٹ آتی ہیں۔",

"what is open merit":
"Open merit is based purely on marks.\nاوپن میرٹ صرف نمبرز پر ہوتا ہے۔",

"what is reserved seat":
"Reserved seats are for specific categories.\nریزرو سیٹ مخصوص لوگوں کے لیے ہوتی ہیں۔",

"how to prepare for entry test":
"Study past papers and practice MCQs.\nپرانے پیپرز پڑھیں اور پریکٹس کریں۔",

"what is interview in admission":
"Interview checks your communication skills.\nانٹرویو آپ کی بات چیت کی صلاحیت چیک کرتا ہے۔",

"do all universities require entry test":
"No, some universities give admission on merit.\nنہیں، کچھ یونیورسٹیاں صرف نمبرز پر داخلہ دیتی ہیں۔",

"how to choose university":
"Choose based on ranking, fees, and location.\nیونیورسٹی کا انتخاب رینکنگ، فیس اور جگہ سے کریں۔",

"what is private university":
"Private university is run by private sector.\nپرائیویٹ یونیورسٹی نجی ادارہ ہوتی ہے۔",

"what is public university":
"Public university is run by government.\nسرکاری یونیورسٹی حکومت چلاتی ہے۔",

"can i apply online":
"Yes, most universities offer online admission.\nجی ہاں، زیادہ تر یونیورسٹیاں آن لائن اپلائی دیتی ہیں۔",

"what is admission deadline":
"Deadline is last date to apply.\nآخری تاریخ اپلائی کرنے کی ڈیڈ لائن ہوتی ہے۔"

})
questions = list(faq_data.keys())
answers = list(faq_data.values())

# -------------------------------
# 🔧 PREPROCESSING
# -------------------------------
def preprocess(text):
    text = text.lower()
    text = "".join([c for c in text if c not in string.punctuation])
    return text

# -------------------------------
# 🤖 RESPONSE ENGINE
# -------------------------------
vectorizer = TfidfVectorizer()

def get_response(user_input):
    user_input = preprocess(user_input)

    all_questions = questions + [user_input]
    tfidf = vectorizer.fit_transform(all_questions)

    similarity = cosine_similarity(tfidf[-1], tfidf[:-1])
    idx = similarity.argmax()
    score = similarity[0][idx]

    if score > 0.2:   # improved threshold
        return answers[idx]
    else:
        return "Sorry, mujhe samajh nahi aaya. Thoda different tareeke se poochain."

# -------------------------------
# 🎤 VOICE INPUT
# -------------------------------
def voice_input():
    if not VOICE_AVAILABLE:
        print("⚠️ Voice feature not available")
        return ""

    r = sr.Recognizer()
    try:
        with sr.Microphone() as source:
            print("🎤 Speak now...")
            r.adjust_for_ambient_noise(source)
            audio = r.listen(source)

        text = r.recognize_google(audio)
        print("You (voice):", text)
        return text

    except Exception as e:
        print("Voice Error:", e)
        return ""

# -------------------------------
# 🔊 VOICE OUTPUT
# -------------------------------
def speak(text):
    if not SPEAK_AVAILABLE:
        return

    try:
        tts = gTTS(text=text, lang='en')
        file = "response.mp3"
        tts.save(file)
        display(Audio(file, autoplay=True))
    except:
        print("Voice output error")

# -------------------------------
# 🔁 MAIN CHAT LOOP
# -------------------------------
def start_chat():
    print("🤖 AI Chatbot Ready!")
    print("Type 'voice' for mic input, 'bye' to exit")

    while True:
        user_input = input("\nYou: ").strip()

        if user_input.lower() == "bye":
            print("Chatbot: Goodbye 👋")
            break

        if user_input.lower() == "voice":
            user_input = voice_input().strip()

        if user_input == "":
            print("Chatbot: Please say something.")
            continue

        response = get_response(user_input)
        print("Chatbot:", response)

        speak(response)

# -------------------------------
# ▶️ RUN
# -------------------------------
start_chat()
```
